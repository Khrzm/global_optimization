```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def func_plot(func_name, color_map):
    lb, ub, dim, fobj = get_functions_details(func_name)
    
    if func_name == 'F5':
        x = np.arange(lb, ub+2, 2)
        y = x

        X, Y = np.meshgrid(x, y)
        Z = np.zeros_like(X)

        for i in range(len(x)):
            for j in range(len(y)):
                Z[i, j] = fobj([x[i], y[j]])

        Z_min = np.min(Z)
        Z_max = np.max(Z)
        Z = 2.0 * (Z - Z_min) / (Z_max - Z_min)

        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        surf = ax.plot_surface(Y, X, Z, cmap=color_map, alpha=0.7)

        # Adjust tick locations
        ax.set_xticks(np.arange(min(y), max(y) + 1, 50))
        ax.set_yticks(np.arange(min(x), max(x) + 1, 50))
        ax.set_zticks(np.arange(0, 2.1, 0.5))

        plt.title(func_name)
        plt.xlabel('x_2')
        plt.ylabel('x_1')
        ax.set_zlabel('Function Value')
        plt.show()
```

```python
def get_functions_details(F):
    lb, ub, dim, fobj = 0, 0, 0, None

    if F == 'F5':
        fobj = F5
        lb = -100
        ub = 100
        dim = 20

    return lb, ub, dim, fobj
```

```python
# Define the objective function F5
def F5(x):
    x = np.array(x)  # Convert x to a NumPy array
    dim = len(x)
    return sum(100 * (x[1:dim] - x[0:dim-1]**2)**2 + (x[0:dim-1] - 1)**2)
```

```python
# Automatically run F5 with a chosen color map (e.g., 'viridis')
color_map = 'viridis'  # Replace 'viridis' with your desired colormap
func_plot('F5', color_map)
```
